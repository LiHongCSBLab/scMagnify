# Regulation Inference and Analysis with scMagnify

## Preliminaries

In this tutorial, you will learn how to:

- **Multi-scale Regulation Inference.** scMagnify builds upon **nonlinear multivariable Granger causal inference**, implemented through an interpretable multi-scale neural network.

- **Network Topology Analysis.** Analyze the GRNs using graph theory to identify critical regulators.

See our manuscript for more details.

### Input

scMagnify uses two types of input data during the regulation inference.

- Input data 1: **scRNA-seq data.** Please look at the previous section to learn about the scRNA-seq data preprocessing and cell state dynamics analysis.

- Input data 2: **Base-GRN.** The base GRN represents the TF-target gene connections. The data structure is a binary 2D matrix or linklist. You can construct the Base-GRN following our workflow, or alternatively, provide a custom one.


### GRNMuData

The `GRNMuData` class extends the `MuData` class to incorporate Gene Regulatory Networks (GRNs) and associated analysis results. Additionally, it provides a set of functions for network pruning and object conversion. Subsequent analyses will be based on this object.

### Load the data

```python
mdata
```

### Train the model

```python
magni = MAGNI(mdata,
             max_iter=1000, 
             patience=10, 
             lag=5, 
             batch_size=32)
```

```python
model = magni.train()
```

```python
plt.plot(model.history["train_loss"])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

### Regulation Inference

```python
gdata = magni.regulation_inference()
```
