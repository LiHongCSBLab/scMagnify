# Base-GRN Construction with scMagnify 

## Preliminaries

scMagnify is a software toolkit to infer gene regulatory networks(GRNs) and explore dynamic regulation synergy from single-cell multiome data.

In this tutorial, you will learn how to:

- **Metacell/Pseudobulk Construction.** Use SEACells {cite} to construct metacells from single-cell multiome data.
- **cCREs Identification** Identify significant peak–gene associations to get cCREs. 
- **Motif Scan.** Following the identification of cCREs, scMagnify scans the DNA sequences of candiate regions to identify transcription factor (TF)-binding motifs. 

### Caution

This tutorial aims to assist users in preprocessing the scATACseq data for scMagnify analysis.For detailed instructions on the usage of SEACells, please refer to the official tutorial.

- `SEACells` documentation: https://github.com/dpeerlab/SEACells/tree/main/notebooks

### Load the data

```python
mdata = scm.read(os.path.join(dirPjtHome, "3_MuData/t-cell-depleted-bm_preprocessed.h5mu"))
```

```python
mdata
```

### Metacell/Pseudobulk construction

```python
meta_mdata = scm.tl.build_metacells_SEACells(mdata)
```

### Conduct lineage-specifc scATACseq analysis

```python
cell_state = "NaiveB"
```

### Peak-gene correlation test

```python
mdata_fil = mdata[mdata["RNA"].obsm["cell_state_masks"][cell_state]].copy()
meta_mdata_fil = meta_mdata[meta_mdata.obsm["cell_state_masks"][cell_state]].copy()
```

```python
mdata_fil = scm.tl.connect_peaks_genes(mdata_fil, 
                                       meta_mdata_fil, 
                                       cor_cutoff=0.1,
                                       pval_cutoff=0.1,
                                       n_jobs=10)
```

### Motif scan

```python
mdata_fil = scm.tl.match_motif(mdata_fil, 
                               motif_db="HOCOMOCOv11_HUMAN")
```

### Save the data

```python
mdata_fil.write()
```
